```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    # Quadratic equation representing temperature change
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    try:
        # Read coefficients from file
        with open("/numbers1.txt", "r") as f:
            content = f.read().strip()
            sets = content.split("\n\n")
            coefficient_sets = []
            for set in sets:
                coefficients = [float(num) for num in set.strip().split("\n")]
                coefficient_sets.append(coefficients)

        # Time values from 0 to 10
        time_values = np.linspace(0, 10, 50)

        # Plot results for each set of coefficients
        plt.figure(figsize=(10, 6))
        for i, coeffs in enumerate(coefficient_sets):
            a, b, c = coeffs
            temperature_coefficients = quadratic_model(time_values, a, b, c)
            plt.plot(time_values, temperature_coefficients, label=f'Set {i+1}: a={a}, b={b}, c={c}')

        plt.xlabel('Time')
        plt.ylabel('Temperature')
        plt.legend()
        plt.title('Weather Modelling with Quadratic Equation (Multiple Sets of Coefficients)')
        #plt.grid(True)  # Add a grid for better visualization
        plt.show()

    except FileNotFoundError:
        print("Error: 'numbers.txt' file not found. Please create it with the coefficients on separate lines.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a=10, b=5, c=-20):
    # Quadratic equation representing temperature change
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    # Time values from 0 to 10
    time_values = np.linspace(0, 10, 50)

    # Calculate temperature using hard-coded coefficients
    a_hardcoded, b_hardcoded, c_hardcoded = 10, 5, -20
    temperature_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)

    # Take input from user for coefficients
    a_user = float(input("Enter Value of a: "))
    b_user = float(input("Enter Value of b: "))
    c_user = float(input("Enter Value of c: "))

    # Calculate temperature using user-defined coefficients
    temp_user = quadratic_model(time_values, a_user, b_user, c_user)

    # Plot results
    plt.plot(time_values, temperature_hardcoded, label=f'Hard coded : a={a_hardcoded}, b={b_hardcoded}, c={c_hardcoded}')
    plt.plot(time_values, temp_user, label=f'User defined : a={a_user}, b={b_user}, c={c_user}', linestyle='--')

    # Annotate the hardcoded coefficients on the plot
    # plt.text(7, quadratic_model(7, a_hardcoded, b_hardcoded, c_hardcoded), f'Hard coded: a={a_hardcoded}, b={b_hardcoded}, c={c_hardcoded}',
             # horizontalalignment='right')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Hard-coded and User-defined Coefficients)')
    plt.show()

if __name__ == "__main__":
    main()
```
